```python
sb_champs = pd.read_excel('Data/Super_Bowl.csv.xlsx')
```

```python
sb_champs.drop(sb_champs[sb_champs['SB'] == 'LVIII (58)'].index, inplace=True)
```

```python
sb_champs['Date'] = pd.to_datetime(sb_champs['Date'])
sb_champs['Year'] = pd.DatetimeIndex(sb_champs['Date']).year
sb_champs = sb_champs[['Year', 'Winner']]
```

```python
sb_champs.loc[:, 'Year'] = sb_champs['Year'] - 1
sb_champs.loc[:, 'Year'] = sb_champs['Year'].astype('object')
sb_champs.loc[:, 'Super_Bowl'] = 1
```

```python
sb_champs.loc[:, 'Year'] = sb_champs['Year'].astype('int')
all_data.loc[:, 'Year'] = all_data['Year'].astype('int')
```

```python
final_df = pd.merge(all_data, sb_champs, left_on = ['Year', 'Team'], right_on = ['Year', 'Winner'], how = 'left')
```

```python
final_df.drop(['Super Bowl', 'Winner'], axis=1, inplace=True)
final_df.loc[:, 'Super_Bowl'] = final_df['Super_Bowl'].fillna(0)
final_df.loc[:, 'Super_Bowl'] = final_df['Super_Bowl'].astype('int')
```
