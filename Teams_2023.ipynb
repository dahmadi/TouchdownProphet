```python
nfl_df = pd.concat([niners, 
                   bears,
                   bengals,
                   bills,
                   broncos,
                   browns,
                   buccaneers,
                   cardinals,
                   chargers,
                   chiefs,
                   colts,
                   commanders,
                   cowboys,
                   dolphins,
                   eagles,
                   falcons,
                   giants,
                   jaguars,
                   jets,
                   lions,
                   packers,
                   panthers,
                   patriots,
                   raiders,
                   rams,
                   ravens,
                   saints,
                   seahawks,
                   steelers,
                   texans,
                   titans,
                   vikings
    ]
)

nfl_df
```

```python
nfl_df = nfl_df.fillna(1)
nfl_df = nfl_df.replace('@', 0)
nfl_df = nfl_df.replace('W', 1)
nfl_df = nfl_df.replace('L', 0)

nfl_df.head()
```

```python
LR = pd.read_csv(('Data/LR_Predictions.csv'), header=0)
ADA = pd.read_csv(('Data/ADA_Predictions.csv'), header=0)
LR.head()
```

```python
LR_prediction = LR[['Team','Prediction']]
ADA_prediction = ADA[['Team', 'Prediction']]

LR_prediction.columns = ['Team','Prediction_LR']
ADA_prediction.columns = ['Team','Prediction_ADA']
```

```python
nfl_predictions_df = pd.merge(nfl_df,
                              LR_prediction,
                              left_on='Team', right_on='Team', how='left'
)

final_nfl_predictions_df = pd.merge(nfl_predictions_df,
                              ADA_prediction,
                              left_on='Team', right_on='Team', how='left'
)
final_nfl_predictions_df.head(25)
```

```python
final_nfl_predictions_df.to_csv('Data/NFL_Predictions_df.csv', index=False)
```
